```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Gravity (m/s^2)
d = 0.002 # Diameter of the circular exit (m)

# Height range (h decreasing from 0.3 m to 0)
h_values = np.linspace(0.3, 0.0000, 390)  # Avoid zero to prevent division errors

# Torricelli's velocity
v_torricelli = np.sqrt(2 * g * h_values)

# Derived velocity equation
denominator = np.sqrt(1 - (81 * d**4) / (256 * h_values**4))
v_your_formula = np.sqrt(2 * g * h_values) / denominator

# Plotting
plt.figure(figsize=(8,5))
plt.plot(h_values, v_torricelli, label="Torricelli's velocity (√2gh)", linestyle="--", color="blue")
plt.plot(h_values, v_your_formula, label="Derived exit velocity", linestyle = "solid", color="orange")
plt.xlabel("Height (h) [m]")
plt.ylabel("Exit Velocity (v) [m/s]")
plt.legend()
#plt.grid()
#plt.xlim(0, 0.01)
#plt.ylim(0, 0.6)
plt.gca().invert_xaxis()  # Invert x-axis for h decreasing
plt.show()
```

```python
# Pressure gradient values
p1 = 0
p2 = 1
p3 = -1

# Normalised height range from 0 to 1
nheight = np.linspace(0,1,100)

# Solved normalised velocity equation
n_vx1 = nheight*(1- ((1/2)*p1*(1-nheight)))
n_vx2 = nheight*(1- ((1/2)*p2*(1-nheight)))
n_vx3 = nheight*(1- ((1/2)*p3*(1-nheight)))

# Plotting the graphs into one
plt.figure(figsize=(8,5))
plt.plot(nheight, n_vx1, label=r"$p_{x}^{'} = 0$", linestyle="solid", color="black")
plt.plot(nheight, n_vx2, label=r"$p_{x}^{'} = 1$", linestyle = "solid", color="red")
plt.plot(nheight, n_vx3, label=r"$p_{x}^{'} = -1$", linestyle = "solid", color="blue")
plt.xlabel(r"$\frac{y}{h}$", fontsize= 15)
plt.ylabel(r"$\frac{v_{x}}{V}$", rotation = 0, fontsize= 15)
plt.legend()
#plt.grid()
plt.show()
```
